#**PERGUNTAS**

---
## 1. Qual é a porcentagem de sobreviventes em relação ao total de passageiros?
---

```python
total_passageiros = len(titanic_df)
sobreviventes = titanic_df['Survived'].sum()
porcentagem_sobreviventes = (sobreviventes / total_passageiros) * 100
print(f"Porcentagem de sobreviventes: {porcentagem_sobreviventes:.2f}%")
```

---
##2. Qual é a classe com a maior quantidade de sobreviventes?
---

```python
classe_sobreviventes = titanic_df.groupby('Pclass')['Survived'].sum().idxmax()
print(f"Classe com a maior quantidade de sobreviventes: {classe_sobreviventes}")
```

---
##3. Qual é a média de tarifa paga pelos passageiros de cada classe?
---

```python
media_tarifa_classe = titanic_df.groupby('Pclass')['Fare'].mean()
print(media_tarifa_classe)
```

---
##4. Qual é a distribuição de idade dos passageiros?
---

```python
distribuicao_idade = titanic_df['Age'].describe()
print(distribuicao_idade)
```

---
##5. Qual é a relação entre a quantidade de irmãos a bordo e a sobrevivência?
---

```python
relacao_sibsp = titanic_df.groupby('SibSp')['Survived'].mean()
print(relacao_sibsp)
```

---
## 6. Qual é a relação entre a quantidade de pais/filhos a bordo e a sobrevivência?
---

```python
relacao_parch = titanic_df.groupby('Parch')['Survived'].mean()
print(relacao_parch)
```

---
##7. Qual é a média de idade dos sobreviventes em comparação com os não sobreviventes?
---

```python
media_idade_sobreviventes = titanic_df[titanic_df['Survived'] == 1]['Age'].mean()
media_idade_nao_sobreviventes = titanic_df[titanic_df['Survived'] == 0]['Age'].mean()
print(f"Média de idade dos sobreviventes: {media_idade_sobreviventes:.2f}")
print(f"Média de idade dos não sobreviventes: {media_idade_nao_sobreviventes:.2f}")
```

---
##8. Quantos passageiros embarcaram em cada porto?
---

```python
contagem_embarque = titanic_df['Embarked'].value_counts()
print(contagem_embarque)
```

---
##9. Qual é o sobrenome mais comum entre os passageiros?
---

```python
titanic_df['Sobrenome'] = titanic_df['Name'].apply(lambda x: x.split(',')[0])
sobrenome_mais_comum = titanic_df['Sobrenome'].value_counts().idxmax()
print(f"Sobrenome mais comum: {sobrenome_mais_comum}")
```

---

---

```python
distribuicao_tarifas = titanic_df['Fare'].describe()
print(distribuicao_tarifas)
```
